## Jednoduchá verze promptu
Nyní, když rozumíme našemu úkolu a typu dat, se kterými pracujeme, začněme vytvářet náš prompt. Začneme s jednoduchou verzí a postupně ji zdokonalíme, abychom zvládli složitosti našeho úkolu shrnutí hovorů.

Začněme s tímto velmi jednoduchým promptem, který načrtává základní úkol:

````python
prompt = """
Summarize the following customer service call transcript. Focus on the main issue, how it was resolved, and any required follow-up.

{transcript}
"""
````

Tento základní prompt dává Claudovi obecnou představu o tom, co chceme, ale má několik omezení:

* Nespecifikuje požadovaný formát výstupu, což by mohlo vést k nekonzistentním shrnutím.
* Neposkytuje pokyny, jak se vypořádat s různými scénáři (jako jsou nevyřešené problémy nebo nedostatečné informace).
* Nestanovuje žádná omezení na délku nebo obsah, což by mohlo vést k příliš dlouhým nebo podrobným shrnutím.
* Neinstrukuje Clauda, aby vynechal osobní informace, což by mohlo vést k problémům s ochranou soukromí.

S tím řečeno, pojďme to otestovat, abychom získali představu o tom, jak to funguje:

````python
```python
from anthropic import Anthropic
from dotenv import load_dotenv

load_dotenv()
client = Anthropic()

def summarize_call(transcript):
    final_prompt = prompt.format(transcript=transcript)
    # Proveď API volání
    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        max_tokens=4096,
        messages=[
            {"role": "user", "content": final_prompt}
        ]
    )
    print(response.content[0].text)
```
````

````python
summarize_call(call1)
````

````python
summarize_call(call2)
````

````python
summarize_call(call3)
````

Jak můžete vidět, i když Claude poskytuje shrnutí, není to ve formátu, který by bylo snadné systematicky analyzovat. Shrnutí může být příliš dlouhé nebo příliš krátké a nemusí konzistentně pokrývat všechny body, které nás zajímají.

V dalších krocích začneme přidávat více struktury a vedení do našeho promptu, abychom tyto omezení řešili. Uvidíme, jak každé přidání zlepšuje kvalitu a konzistenci shrnutí od Claude.

## Strukturování našeho hlavního promptu

### Instrukce a formát výstupu

Než budeme pokračovat, pojďme si jasně promyslet, jak by mohl vypadat dobře strukturovaný výstupní formát. Abychom si usnadnili práci při zpracování výsledků, je často nejjednodušší požádat Claude o odpověď ve formátu JSON. Jak by měl v tomto případě vypadat dobrý JSON?

Minimálně by náš JSON výstup měl obsahovat následující:
- Stav, zda měl Claude dostatek informací k vytvoření shrnutí. K tomu se ještě vrátíme. Prozatím budeme předpokládat, že všechna shrnutí mají stav "COMPLETE", což znamená, že Claude mohl vytvořit shrnutí.
- Shrnutí problému zákazníka
- Zda hovor vyžaduje další následné kroky
- Podrobnosti o jakýchkoli následných akcích, pokud jsou vyžadovány (zavolat zákazníkovi zpět atd.)
- Jak byl problém vyřešen
- Seznam nejasností nebo vágních bodů v konverzaci

Zde je navrhovaná ukázková struktura JSON:

```json
{
  "summary": {
    "customerIssue": "Stručný popis hlavního problému nebo důvodu hovoru",
    "resolution": "Jak byl problém řešen nebo vyřešen, pokud je to relevantní",
    "followUpRequired": true/false,
    "followUpDetails": "Popis jakýchkoli nezbytných následných akcí, nebo null, pokud nejsou vyžadovány"
  },
  "status": "COMPLETE",
  "ambiguities": ["Seznam jakýchkoli nejasných nebo vágních bodů v konverzaci, nebo prázdné pole, pokud žádné nejsou"]
}
```

Vytvořme novou část našeho promptu, která zahrnuje konkrétní instrukce, včetně:
- Vytvoření shrnutí zaměřeného na hlavní problém, řešení a jakékoli požadované následné kroky.
- Generování výstupu ve formátu JSON podle našeho specifického, standardizovaného formátu.
- Vynechání konkrétních informací o zákazníkovi ve shrnutích.
- Udržení každé části shrnutí krátké.

Zde je pokus o poskytnutí instrukcí pro výstup, včetně našeho specifického formátu výstupu JSON:

````python
prompt_pt2 = """
Instructions:
1. Read the transcript carefully.
2. Analyze the transcript, focusing on the main issue, resolution, and any follow-up required.
3. Generate a JSON object summarizing the key aspects of the interaction according to the specified structure.

Important guidelines:
- Confidentiality: Omit all specific customer data like names, phone numbers, and email addresses.
- Character limit: Restrict each text field to a maximum of 100 characters.
- Maintain a professional tone in your summary.

Output format:
Generate a JSON object with the following structure:
<json>
{
  "summary": {
    "customerIssue": "Brief description of the main problem or reason for the call",
    "resolution": "How the issue was addressed or resolved, if applicable",
    "followUpRequired": true/false,
    "followUpDetails": "Description of any necessary follow-up actions, or null if none required"
  },
  "status": "COMPLETE",
  "ambiguities": ["List of any unclear or vague points in the conversation, or an empty array if none"]
}
</json>
"""
````

## Použití XML značek a poskytnutí prostoru pro přemýšlení Claudovi
Dále použijeme dvě další strategie promptování: poskytnutí prostoru pro přemýšlení Claudovi a použití XML značek.
- Požádáme Clauda, aby začal výstupem značek `<thinking>`, které obsahují jeho analýzu.
- Poté požádáme Clauda, aby svůj JSON výstup vložil do `<json>`.

Zde je závěrečná část našeho prvního návrhu promptu:

````python
prompt_pt3 = """
Before generating the JSON, please analyze the transcript in <thinking> tags. 
Include your identification of the main issue, resolution, follow-up requirements, and any ambiguities. 
Then, provide your JSON output in <json> tags.
"""
````

Tím, že požádáme Claude, aby vložil svou analýzu do značek `<thinking>`, ho vybízíme k tomu, aby rozložil svůj myšlenkový proces před formulováním konečného JSON výstupu. To podporuje důkladnější a strukturovanější přístup k analýze přepisu. Sekce `<thinking>` nám (a potenciálně dalším recenzentům nebo systémům) umožňuje vidět Claudeův proces uvažování. Tato transparentnost může být klíčová pro účely ladění a zajištění kvality.

Oddělením analýzy (`<thinking>`) od strukturovaného výstupu (`<json>`) vytváříme jasné rozlišení mezi Claudeovou interpretací přepisu a jeho formátovaným shrnutím. To může být užitečné v případech, kdy bychom chtěli přezkoumat analýzu odděleně od JSON výstupu, ale také tím, že izolujeme JSON obsah uvnitř značek `<json>`, usnadňujeme analýzu konečné odpovědi a zachycení JSON, se kterým chceme pracovat.

## Další vylepšení promptu

Jak jsme již viděli, náš prompt aktuálně generuje úplné shrnutí pro okrajové případy přepisů. Chceme toto chování změnit. Máme několik možností, jak tyto okrajové případy řešit:

- Označit je nějakým způsobem, aby bylo zřejmé, že nejsou shrnutelné, což umožní pozdější lidskou kontrolu.
- Kategorizovat je zvlášť (např. "technické potíže", "jazyková bariéra" atd.).

Pro zjednodušení se rozhodneme tyto okrajové případy označit tím, že požádáme model, aby vygeneroval JSON, který vypadá takto:

```json
{
  "status": "INSUFFICIENT_DATA"
}
```
