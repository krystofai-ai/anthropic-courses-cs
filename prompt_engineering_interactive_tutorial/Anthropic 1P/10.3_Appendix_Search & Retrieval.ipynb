# Dodatek 10.3: Vyhledávání a získávání informací

Věděli jste, že můžete použít Claude k **vyhledávání na Wikipedii za vás**? Claude dokáže najít a získat články, a poté můžete také použít Claude k jejich shrnutí a syntéze, psaní nového obsahu z toho, co našel, a mnohem více. A nejen Wikipedii! Můžete také prohledávat své vlastní dokumenty, ať už uložené jako prostý text nebo vložené ve vektorové databázi.

Podívejte se na naše [příklady z RAG kuchařky](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb), abyste se naučili, jak doplnit znalosti Claude a zlepšit přesnost a relevanci odpovědí Claude pomocí dat získaných z vektorových databází, Wikipedie, internetu a dalších zdrojů. Tam se také můžete dozvědět, jak používat určité [embeddings](https://docs.anthropic.com/claude/docs/embeddings) a nástroje pro vektorové databáze.

Pokud máte zájem o učení se o pokročilých RAG architekturách s použitím Claude, podívejte se na naše [technické prezentační slajdy Claude 3 o RAG architekturách](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
