## Nastavení

Spusťte následující buňku nastavení pro načtení vašeho API klíče a vytvoření pomocné funkce `get_completion`.

````python
%pip install anthropic --quiet

# Importujte modul hints z balíčku utils
import os
import sys
module_path = ".."
sys.path.append(os.path.abspath(module_path))
from utils import hints

# Importujte vestavěnou knihovnu regulárních výrazů v Pythonu
import re
from anthropic import AnthropicBedrock

%store -r MODEL_NAME
%store -r AWS_REGION

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
````

## Lekce

**Claude reaguje nejlépe na jasné a přímé instrukce.**

Představte si Claude jako jakéhokoli jiného člověka, který je nový v práci. **Claude nemá žádný kontext** o tom, co dělat, kromě toho, co mu doslova řeknete. Stejně jako když poprvé instruujete člověka o úkolu, čím více přesně vysvětlíte, co chcete, přímočaře Claude, tím lepší a přesnější bude Claudeova odpověď.

### Příklady

Podívejme se na úkol, jako je psaní poezie. (Ignorujte jakýkoli nesoulad v počtu slabik - LLM zatím nejsou skvělé v počítání slabik.)

````python
```python
# Prompt
PROMPT = "Write a haiku about robots."

# Vytiskne odpověď od Claude
print(get_completion(PROMPT))
```
````

Toto haiku je dost pěkné, ale uživatelé mohou chtít, aby Claude přešel přímo k básni bez úvodu "Zde je haiku".

Jak toho dosáhneme? **Požádáme o to!**

````python
```python
# Prompt
PROMPT = "Write a haiku about robots. Skip the preamble; go straight into the poem."

# Vytisknout odpověď od Claude
print(get_completion(PROMPT))
```
````

Zde je další příklad. Zkusme se zeptat Claude, kdo je nejlepší basketbalista všech dob. Jak můžete vidět níže, zatímco Claude uvádí několik jmen, **neodpovídá s definitivním "nejlepším"**.

````python
```python
# Prompt
PROMPT = "Who is the best basketball player of all time?"

# Vytiskne odpověď od Claude
print(get_completion(PROMPT))
```
````

Můžeme přimět Claude, aby se rozhodl a vybral nejlepšího hráče? Ano! Stačí se zeptat!

````python
```python
# Prompt
PROMPT = "Who is the best basketball player of all time? Yes, there are differing opinions, but if you absolutely had to pick one player, who would it be?"

# Vytisknout Claudeovu odpověď
print(get_completion(PROMPT))
```
````

### Cvičení 2.1 - Španělština
Upravte `SYSTEM_PROMPT`, aby Claude poskytl svou odpověď ve španělštině.

````python
```python
# Systémový prompt - toto je jediné pole, které byste měli změnit
SYSTEM_PROMPT = "[Replace this text]"

# Prompt
PROMPT = "Hello Claude, how are you?"

# Získat odpověď od Claude
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Funkce pro hodnocení správnosti cvičení
def grade_exercise(text):
    return "hola" in text.lower()

# Vytisknout odpověď od Claude a odpovídající hodnocení
print(response)
print("\n--------------------------- HODNOCENÍ ---------------------------")
print("Toto cvičení bylo správně vyřešeno:", grade_exercise(response))
```
````

### Cvičení 2.2 - Pouze jeden hráč

Upravte `PROMPT` tak, aby Claude vůbec neváhal a odpověděl **POUZE** jménem jednoho konkrétního hráče, **bez dalších slov nebo interpunkce**.

````python
```python
# Prompt - toto je jediné pole, které byste měli změnit
PROMPT = "[Replace this text]"

# Získání odpovědi od Claude
response = get_completion(PROMPT)

# Funkce pro hodnocení správnosti cvičení
def grade_exercise(text):
    return text == "Michael Jordan"

# Tisk odpovědi od Claude a odpovídající hodnocení
print(response)
print("\n--------------------------- HODNOCENÍ ---------------------------")
print("Toto cvičení bylo správně vyřešeno:", grade_exercise(response))
```
````

### Cvičení 2.3 - Napište příběh

Upravte `PROMPT` tak, aby Claude odpověděl co nejdelší odpovědí, jakou dokážete vytvořit. Pokud vaše odpověď bude **přes 800 slov**, bude Claudeova odpověď hodnocena jako správná.

````python
```python
# Prompt - toto je jediné pole, které byste měli změnit
PROMPT = "[Replace this text]"

# Získat odpověď od Claude
response = get_completion(PROMPT)

# Funkce pro hodnocení správnosti cvičení
def grade_exercise(text):
    trimmed = text.strip()
    words = len(trimmed.split())
    return words >= 800

# Vytisknout odpověď od Claude a odpovídající hodnocení
print(response)
print("\n--------------------------- HODNOCENÍ ---------------------------")
print("Toto cvičení bylo správně vyřešeno:", grade_exercise(response))
```
````
