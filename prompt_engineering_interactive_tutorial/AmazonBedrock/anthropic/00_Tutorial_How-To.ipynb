### Anthropic SDK a Messages API
V průběhu tohoto tutoriálu budeme používat [Anthropic python SDK](https://docs.anthropic.com/claude/reference/claude-on-amazon-bedrock) a [Messages API](https://docs.anthropic.com/claude/reference/messages_post).

Níže je příklad, jak bude vypadat spuštění promptu v tomto tutoriálu.

Nejprve nastavíme a uložíme název modelu a region.

````python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
````

Poté vytvoříme `get_completion`, což je pomocná funkce, která odešle prompt do Claude a vrátí generovanou odpověď od Claude. Nyní spusťte tuto buňku.

````python
from anthropic import AnthropicBedrock

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
````

Nyní napíšeme příklad promptu pro Claude a vytiskneme výstup Claude spuštěním naší pomocné funkce `get_completion`. Spuštění buňky níže vytiskne odpověď od Claude pod ní.

Neváhejte si pohrát s řetězcem promptu, abyste získali různé odpovědi od Claude.

````python
```python
# Prompt
prompt = "Hello, Claude!"

# Získat odpověď od Claude
print(get_completion(prompt))
```
````
