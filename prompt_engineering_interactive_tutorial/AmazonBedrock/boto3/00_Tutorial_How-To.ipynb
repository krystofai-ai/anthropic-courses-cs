### Anthropic SDK a Messages API
V průběhu tohoto tutoriálu budeme používat [Anthropic python SDK](https://docs.anthropic.com/claude/reference/client-sdks) a [Messages API](https://docs.anthropic.com/claude/reference/messages_post).

Níže je příklad, jak bude vypadat spuštění promptu v tomto tutoriálu. Nejprve vytvoříme `get_completion`, což je pomocná funkce, která odešle prompt do Claude a vrátí generovanou odpověď od Claude. Spusťte tuto buňku nyní.

Nejprve nastavíme a uložíme název modelu a region.

````python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
````

Poté vytvoříme `get_completion`, což je pomocná funkce, která odešle prompt do Claude a vrátí generovanou odpověď od Claude. Nyní spusťte tuto buňku.

````python
import boto3
import json

bedrock = boto3.client('bedrock-runtime',region_name=AWS_REGION)

def get_completion(prompt):
    body = json.dumps(
        {
            "anthropic_version": '',
            "max_tokens": 2000,
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.0,
            "top_p": 1,
            "system": ''
        }
    )
    response = bedrock.invoke_model(body=body, modelId=MODEL_NAME)
    response_body = json.loads(response.get('body').read())

    return response_body.get('content')[0].get('text')
````

Nyní napíšeme příklad promptu pro Claude a vytiskneme výstup Claude spuštěním naší pomocné funkce `get_completion`. Spuštění buňky níže vytiskne odpověď od Claude pod ní.

Neváhejte si pohrát s řetězcem promptu, abyste získali různé odpovědi od Claude.

````python
```python
# Prompt
prompt = "Hello, Claude!"

# Získání odpovědi od Claude
print(get_completion(prompt))
```
````

Proměnné `MODEL_NAME` a `AWS_REGION`, které byly definovány dříve, budou použity v celém tutoriálu. Jen se ujistěte, že spouštíte buňky pro každou stránku tutoriálu odshora dolů.
